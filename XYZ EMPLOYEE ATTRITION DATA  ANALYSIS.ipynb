```python
print('DATA STATISTIC BY CHI-SQUARE STATISTIC')
```

```python
from scipy import stats

# Chi-square test for categorical variables
contingency_table = pd.crosstab(data['Department'], data['Attrition'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}, P-value: {p}")
```

```python
# Attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=data)
plt.title('Attrition by Department')
plt.show()

#  Attrition by Age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='Age', data=data)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
print('DATA VISUALIZATION BY PIE CHART. Pie chart is used to present the data to that count the occurances of each unique value in the attrition column, which provides the number of employees who left(YES) and the number of employees who stayed(NO).')
```

```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the Employee Attrition Dataset
# Replace 'employee_data.csv' with the path to your actual dataset
data = pd.read_csv('attrition data.csv')

# Step 3: Prepare the Data for the Pie Chart
# Count the number of employees who left vs those who stayed
attrition_counts = data['Attrition'].value_counts()

# Step 4: Create the Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(
    attrition_counts,
    labels=attrition_counts.index,
    autopct='%1.1f%%',  # Display percentages
    startangle=140,      # Start angle for the pie chart
    colors=['#ff9999','#66b3ff']  # Optional: specify colors for the pie slices
)

# Adding title
plt.title('Employee Attrition Distribution')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
